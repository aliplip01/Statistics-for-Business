```python
# Import necessary library

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

Dataset source: 'https://raw.githubusercontent.com/CarrenP/files/refs/heads/main/penguins_size.csv'

```python
# import the df from the link above
URL = 'https://raw.githubusercontent.com/CarrenP/files/refs/heads/main/penguins_size.csv'
df = pd.read_csv(URL)
df.head()
```

```python
# menunjukan informasi mengenai tabel, dari data type, jumlah data hingga nama kolum.
df.info()
```

```python
# Count the penguins based on their species
df_spesies = df.groupby('species')['island'].count()
print(df_spesies)
```

```python
# create a bar chart to visualise the spesies, explain your findng in at least 2 setences.

df.groupby('island')['species'].count().plot(
    kind = 'pie',y='Island',figsize=(8,6),autopct ='%1.2f%%',legend = True
)

plt.title("Penguin Distribution by Island")
plt.ylabel("")
plt.show()

# berdasarkan hasil pie chart di bawah, diketahui bahwa pinguin di daerah Biscoe adalah yang terbanyak sekitar 48.84%, lalu di daerah Dream sebesar 36.05%.
# dan yang paling sedikit berada di daerah Torgersen yaitu sekitar 15.12%
```

```python
# Display the data distribution of culmen length by island using box plot,explain your findng in at least 2 setences.
df.boxplot(by='island',column=['culmen_length_mm'],grid = False)
plt.show()
```
