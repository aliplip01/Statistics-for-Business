```python
import matplotlib.pyplot as plt
import pandas as pd

data=pd.read_csv('https://raw.githubusercontent.com/CarrenP/files/main/NY-House-Dataset.csv')
data.head()
```

```python
#Create a bar chart to illustrate the Distribution of Houses by Sublocality in New York. explain your finding.
sublocality_counts = data['SUBLOCALITY'].value_counts()
sublocality_counts.plot(kind='bar', color='skyblue')

plt.title('Distribution of Houses by Sublocality in New York')
plt.xlabel('Sublocality')
plt.ylabel('Number of Houses')
plt.show()
```

```python
# Calculate the range of house prices in the dataset.
min_price = data['PRICE'].min()
max_price = data['PRICE'].max()
price_range = max_price - min_price

print(f"Maximum Price: ${max_price}")
print(f"Minimum Price: ${min_price}")
print(f"Range of House Prices: ${price_range}")
```

```python
# Find the standard deviation of the house prices to understand the variability in pricing.
std = data['PRICE'].std()
print(f"Standard Deviation of House Prices: ${std}")
```

```python
# Calculate the average market price of the houses in the dataset.
average = data['PRICE'].mean()
print(f'Average Market Price of Houses: ${average}')
```

```python
# Determine the mode for the number of bedrooms in the dataset.
mode = data['BEDS'].mode()
print(f'Mode for the Number of Bedrooms: {mode[0]}')
```
