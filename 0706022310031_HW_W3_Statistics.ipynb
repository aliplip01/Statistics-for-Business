## Tuple Questions

Create a tuple called midterm_scores that contains the following elements: 80, 90, 65, "incomplete", 78, 85, 95, 100, 85, 70. How do you access the third element in the tuple?

```python
midterm_scores = (80, 90, 65, "incomplete", 78, 85, 95, 100, 85, 70)

third_element = midterm_scores[2]
print(third_element)
```

Access the last three elements of the tuple midterm_scores using slicing.

```python
last_three_elements = midterm_scores[-3:]
print(last_three_elements)
```

Given the tuple temperatures = (30, 32, 35, 33, 31, 29), find the highest temperature using a Python function.

```python
temperatures = (30, 32, 35, 33, 31, 29)

highest_temperature = max(temperatures)
print("The highest temperature is", highest_temperature)
```

Combine two tuples even_numbers = (2, 4, 6) and odd_numbers = (1, 3, 5) into a single tuple.

```python
even_numbers = (2,4,6)
odd_numbers = (1,3,5)

combined_numbers = (even_numbers + odd_numbers)
print(combined_numbers)
```
