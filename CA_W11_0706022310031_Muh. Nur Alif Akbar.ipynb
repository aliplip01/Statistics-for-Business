```python
# Import library
import numpy as np
import pandas as pd
import scipy.stats as st
```

```python
# function to retrieve value from z-table
def get_z_table(alpha, tail):
  if tail == 1:
    return st.norm.ppf(1 - alpha)
  elif tail == 2:
    return st.norm.ppf(1 - (alpha / 2))

# function to retrieve value from t-table
def get_t_table(alpha, tail, dof):
  if tail == 2:
    alpha /= 2  
  return st.t.isf([alpha], [dof])[0]
```

## Case #1A: Two-Tailed Hypothesis Testing, σ Known
Jamestown Steel Company manufactures and assembles desks and other office equipment at several plants in New York State. At the Fredonia plant, the weekly production of the Model A325 desk follows a normal distribution with a mean of 200 and a standard deviation of 16. 
New production methods have been introduced and the vice president of manufacturing would like to investigate whether there has been a <u>change</u> in weekly production of the Model A325. Is the mean number of desks produced different from 200 at the 0.01 significance level?

```python
# Step 1: State null and alternate hypothesis
# H0: μ == 200
# H1: μ != 200

# Step 2: Select level of significance
alpha = 0.01
tail = 2
Xbar = 203.5
Miu = 200
std_dev_pop = 16
n = 50

# Step 3: Identify the test statistics
# Use Z-test because the population standard deviation is known

# Step 4: Formulate a decision rule
# Reject H0 if |z_calc| > z_table(alpha/2)

# Step 5: Take a sample, arrive at decision
z_calc = (Xbar-Miu) / (std_dev_pop / np.sqrt(n))
z_table = get_z_table(alpha, tail)
print("z_calc = {:.4f}".format(z_calc))
print("z_table = {:.4f}".format(z_table))

# Step 6: Interpret the result
if np.abs(z_calc) > z_table:
  print("H0 rejected. The production rate is different from 200")
else:
  print("Fail to Reject H0. The production rate is not different from 200")
```

## Case #1B: One-Tailed Hypothesis Testing, σ Known
Suppose instead of wanting to know if there had been a change in the mean number of desks assembled,  the vice president wanted to know if there had been an <u>increase</u> in the number of units assembled. Can we conclude, because of the improved production methods, that the mean number of desks assembled in the last 50 weeks was more than 200? Please use α = 0.01.

```python
# Step 1: State null and alternate hypothesis
# H0: μ <= 200
# H1: μ > 200

# Step 2: Select level of significance
alpha = 0.01
tail = 1
Xbar = 203.5
Miu = 200
std_dev_pop = 16
n = 50

# Step 3: Identify the test statistics
# Use Z-test because the population standard deviation is known

# Step 4: Formulate a decision rule
# Reject H0 if z_calc > z_table(alpha)

# Step 5: Take a sample, arrive at decision
z_calc = (Xbar-Miu) / (std_dev_pop / np.sqrt(n))
z_table = get_z_table(alpha, tail)
print("z_hitung = {:.2f}".format(z_calc))
print("z_tabel = {:.2f}".format(z_table))

# Step 6: Interpret the result
if z_calc > z_table:
  print("H0 rejected. The production rate is more than 200")
else:
  print("Fail to Reject H0. The production rate is not more than 200")
```

## Case #2: Hypothesis Testing, σ Unknown

image.png

```python
# Step 1: State null and alternate hypothesis
# H0: μ >= 60
# H1: μ < 60

# Step 2: Select level of significance
claim = [45, 48, 48, 58, 38, 49, 53, 54, 51, 76, 62, 67, 51, 58, 40, 63, 56, 59, 43, 78, 63, 56, 61, 64, 69, 57]
alpha = 0.01
tail = 1
Xbar = np.mean(claim)
Miu = 60
# For population standard deviation, delta degree of freedom (ddof) is 0
# For sample standard deviation, delta degree of freedom (ddof) is 1
std_dev_sample = np.std(claim, ddof=1)
n = len(claim)
dof = n - 1

# Step 3: Identify the test statistics
# Use t-test because population standard deviation is unknown

# Step 4: Formulate a decision rule
# Reject H0 if t_calc < -t_table(alpha, dof)


# Step 5: Take a sample, arrive at decision
t_calc = (Xbar-Miu) / (std_dev_sample / np.sqrt(n))
t_table = get_t_table(alpha, tail, dof)
print("t_calc = {:.3f}".format(t_calc))
print("t_table = {:.3f}".format(t_table))

# Step 6: Interpret the result
if t_calc < (-1 * t_table):
  print("H0 rejected. The cost to process a claim is now less than $60")
else:
  print("Fail to Reject H0. The cost to process a claim is still $60 or more")
```

## Exercise 1: Soda Filler Machine

```python
import numpy as np
from scipy.stats import norm

# Step 1: State null and alternate hypothesis
# H0: μ_soda >= 355 ml
# H1: μ_soda < 355 ml

# Step 2: Select level of significance
alpha = 0.05
Xbar = 354.5  # Sample mean
Miu = 355  # Hypothesized mean
std_dev_pop = 3  # Population standard deviation
n = 40  # Sample size

# Step 3: Identify the test statistics
# Use Z-test because the population standard deviation is known


# Step 4: Formulate a decision rule
# Reject H0 if z_calc > z_table(alpha)

# Step 5: Take a sample, arrive at decision
z_calc = (Xbar - Miu) / (std_dev_pop / np.sqrt(n))
z_table = norm.ppf(alpha)

print("z_calc = {:.4f}".format(z_calc))
print("z_table = {:.4f}".format(z_table))

# Step 6: Interpret the result
if z_calc < z_table:
    print("H0 rejected. The average fill is less than 355 ml.")
else:
    print("Fail to Reject H0. The average fill is not significantly less than 355 ml.")
```

## Exercise 3: Coffee Shop Revenue

```python
import numpy as np
from scipy.stats import t

# Step 1: State null and alternate hypothesis
# H0: μ <= 500
# H1: μ > 500

# Step 2: Select level of significance
alpha = 0.05
Xbar = 525  # Sample mean
Miu = 500  # Hypothesized mean
std_dev_sample = 50  # Sample standard deviation
n = 12  # Sample size
dof = n - 1  # Degrees of freedom

# Step 3: Identify the test statistics
# Use T-test because the population standard deviation is unknown

# Step 4: Formulate a decision rule
# Reject H0 if t_calc > t_table(alpha, dof)

# Step 5: Take a sample, arrive at decision
t_calc = (Xbar - Miu) / (std_dev_sample / np.sqrt(n))
t_table = t.ppf(1 - alpha, dof)

print("t_calc = {:.4f}".format(t_calc))
print("t_table (critical value) = {:.4f}".format(t_table))

# Step 6: Interpret the result
if t_calc > t_table:
    print("H0 rejected. The average daily revenue is significantly higher than $500.")
else:
    print("Fail to Reject H0. There is insufficient evidence to conclude that the average daily revenue is higher than $500.")
```
