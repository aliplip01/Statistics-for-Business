### Mean, Median, Modus

image.png

```python
# Buatlah list data_harga_rumah
a = 2000000
b = 500000
c = 300000
d = 100000
e = 100000

n = 5

# Hitunglah mean, median, modus, dan standar deviasi harga rumah
mean = (a + b + c + d + e) / n 
print("Mean = ",mean)

values = [a,b,c,d,e]
sorted_values = sorted(values)
index = len(sorted_values)
if n % 2 == 1:
    median = sorted_values[index // 2]
else:
    median = (sorted_values[index // 2 - 1] + sorted_values[n // 2]) / 2
print("Median = ", median)

frequency = {}

for value in values:
    if value in frequency:
        frequency[value] += 1
    else:
        frequency[value] = 1

max_frequency = max(frequency.values())
modes = [key for key, value in frequency.items() if value == max_frequency]
print("Modes = ",modes)

standar_deviasi = np.std(values)
print("Standar deviasi harga rumah adalah", standar_deviasi)
```

```python
# Buatlah visualisasi harga rumah dengan bar chart
from matplotlib.ticker import FuncFormatter
import matplotlib.pyplot as plt

# Data to be visualized
categories = ['a', 'b', 'c', 'd', 'e']  # Labels for each bar
values = [2000000, 500000, 300000, 100000, 100000]  # Values for each category

def y_axis_format(x, pos):
    return f'{x:,}'

# Create the bar chart
plt.bar(categories, values, color='skyblue')

# Add title and labels
plt.title('Harga Rumah')
plt.xlabel('Nama Rumah')
plt.ylabel('Harga ($)')

# Show the plot
plt.gca().yaxis.set_major_formatter(FuncFormatter(y_axis_format))
plt.show()

```

### Weighted Mean

Sebuah perusahaan konstruksi memiliki skema upah 16.50, 19.00, or 26.00 per jam. Total terdapat 26 pekerja, 14 orang dibayar dengan upah 16.50, 10 orang dengan upah 19.00, dan 2 orang dengan upah 26.00. Berapakah rata-rata upah pekerja?

```python
# Buatlah variabel upah_pekerja dengan type list

upah_pekerja = [16.50] * 14 + [19.00] * 10 + [26.00] * 2
rata_rata_upah = sum(upah_pekerja) / len(upah_pekerja)
print("Rata rata upah pekerja = ", rata_rata_upah)
```

### Geometric Mean

Misalkan Anda menerima kenaikan gaji 5 persen tahun ini dan kenaikan 15 persen tahun depan. Peningkatan persen tahunan rata-rata adalah 9.886%, bukan 10.00%.

```python
# Buatlah list data_kenaikan_gaji

data_kenaikan_gaji = [0.05, 0.15] 

# Gunakan library scipy.stats atau statistics untuk menghitung geometric mean, tampilkan hingga 4 desimal

geometric_mean = stat.geometric_mean([1 + roi for roi in data_kenaikan_gaji])

print(f"Rata-rata kenaikan gaji adalah: {geometric_mean:.4f}")
```

The return on investment earned by Atkins
construction Company for four successive years was:
30 percent, 20 percent, - 40 percent, and 200
percent. What is the geometric mean rate of return
on investment?

```python
import statistics as stat

# List data ROI (persentase)
data_roi = [30, 20, -40, 200]

# Menghitung geometric mean
geometric_mean_roi = stat.geometric_mean([1 + roi / 100 for roi in data_roi])

# Menghitung dan menampilkan hasil dalam persen
print(f"Geometric Mean Rate of Return: {geometric_mean_roi:.4f}")
```

Selama dekade 1990an, dan memasuki tahun 2000an, Las Vegas, Nevada, adalah kota dengan pertumbuhan tercepat di AS. Populasi meningkat dari 258.295 pada tahun 1990 menjadi 607.876 pada tahun 2009. Ini meningkat 349.581 orang atau meningkat 135,3% dari periode tersebut. Populasi lebih dari dua kali lipat.  Berapakah kenaikan tahunan rata-rata?
image.png

```python
# Gunakan rumus kenaikan tahunan rata-rata
from scipy.stats import gmean

# Populasi awal dan akhir
P = 258295  # Populasi pada tahun 1990
A = 607876  # Populasi pada tahun 2009
n = 2009 - 1990  # Jumlah tahun

# Menghitung Compound Annual Growth Rate (CAGR)
CAGR = (A / P) ** (1 / n) - 1

# Menampilkan CAGR dalam bentuk persentase
print(f"CAGR (Rate of Increase Over Time): {CAGR:.4f}")

# Geometric mean (menghitung pertumbuhan tahunan rata-rata)
# Untuk Geometric Mean, kita akan menggunakan faktor pertumbuhan tahunan
growth_factors = [(A / P) ** (1 / n)]  # Faktor pertumbuhan tahunan konstan
geometric_mean = gmean(growth_factors) - 1

# Menampilkan Geometric Mean dalam bentuk persentase
print(f"Geometric Mean: {geometric_mean:.4f}")
```
