#### Additional Challenge #1 (Optional)
The camelCase convention is a widely used variable name practise in which each word within a compound word is capitalised, with the exception of the initial word. CamelCase is a commonly employed convention by software developers for naming variables in source code. Some examples of camelCase in programming are variables such as "myArray" and "getCommandException." The objective is to transform a given string input into an output string in camelCase format.

For example:

*   input: "I am a Proud ISB Students"
*   output: "iAmAProudIsbStudents"

hint: use split(), lower(), capitalize(), and join() function

```python
def to_camel_case(input_string):
    txt = input_string.split()
    return txt[0].lower() + ''.join(txt.capitalize() for txt in txt[1:])
    return camel_case_string

input_string = 'I am a Proud ISB Students'
camel_case_result = to_camel_case(input_string)
print(camel_case_result)
```

#### Additional Challenge #2 (Optional)
Encryption is the process of encoding a message or file so that it can only be deciphered by authorised individuals. Encryption employs an algorithm to scramble, or encrypt, data and a key to decipher, or decrypt, the information for the receiving party. Your mission is to generate a straightforward encryption of the String by shifting its Unicode number.

For example:

*   input: "I am a proud ISB student"
*   shift by: 1 unicode
*   output: "J!bn!b!qspve!JTC!tuvefou"

hint: use chr() and ord() function

```python
def my_encryption(my_input):
    decrypted = ''
    for a in my_input :
        decrypted += chr(ord(a) + 1)
    return decrypted

input_string = 'I am a Proud ISB Students'
encryption_result = my_encryption(input_string)
print(encryption_result)
```
